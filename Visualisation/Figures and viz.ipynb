```python
import matplotlib.pyplot as plt
```

```python
x = 5
y = 5

plt.plot(x,y,'gp')
plt.axis('square')
plt.axis([-6,6 ,-6,6])
plt.grid()
plt.show()
```

```python
x = [3,4,5]
y = [0,4,3]

for i in range(0,len(x)):
    plt.plot(x[i], y[i], 'o', label = ' point %s' %(i+1))
    
plt.legend()    
plt.show()
```

```python
# y = x**2 - 3*x
for xi in range(-10, 11):
    yi = xi**2 -3*xi
    plt.plot(xi, yi, 'o')
    
plt.show()
```

```python
import sympy as sym 

x = sym.symbols('x')
y = x**2 - 3*x

for pi in range(-11,12):
    plt.plot(pi, y.subs(x, pi) ,'o')
plt.show()
```

```python
p1 = [-3, 1]
p2 = [4,4]

plt.plot([p1[0], p2[0]], [p1[1], p2[1]], color =[.4,.2,.7], linewidth = 3)

plt.axis('square')
plt.grid()
plt.show()
```

```python
import numpy as np 
x = range(-20,21)

for i in x:
    plt.plot([0,i], [0, np.sqrt(abs(i))])

plt.show()
```

```python
plt.plot([0,2], [2,2], 'r')
plt.plot([0,2], [0,0], 'r')
plt.plot([0,0], [0,2], 'r')
plt.plot([2,2], [0,2], 'r')

plt.axis('square')
plt.axis([-3,5, -3,5])
plt.show()
```

```python
x= range(-3, 4)
y = np.zeros(len(x))

for i in range (0 , len(x)):
    y[i] = 2 - x[i]**2
    
plt.plot(x,y, 's-')
```

```python
x = np.linspace(-3, 3, 20)

y = 2 + np.sqrt(abs(x))


plt.plot(x,y, 'rs-')
```

```python
import sympy.plotting.plot as symplot
x, a = sym.symbols('x,a')
expr = a/x
n=1
p = symplot(expr.subs(a,n), (x,-5,5), show = False)
p[0].label = '$y=%s$' %sym.latex(expr.subs(a,n))

p2 = symplot(expr.subs(a,2), show = False)
p.extend(p2)

p[0].line_color = 'r'

p.ylim = [-5, 5]
p.legend = True
p.show()
```
