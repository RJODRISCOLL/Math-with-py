A harmonograph is a mechanical apparatus that employs pendulums to create a geometric image. The drawings created typically are Lissajous curves or related drawings of greater complexity. The devices, which began to appear in the mid-19th century and peaked in popularity in the 1890s, cannot be conclusively attributed to a single person, although Hugh Blackburn, a professor of mathematics at the University of Glasgow, is commonly believed to be the official inventor. (WIKIPEDIA) 







```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# initial params 
n = 1000000
t = np.logspace(np.log10(10),np.log10(500),n)

# parameters 
A = [  1, 1, 1.5, 1.5 ]
d = [ .006, .0005, .004, .0022 ]
f = [   3, 1, 1, 2.5 ]

# generate pairs
# in this formula we have some sin waves that are tapered by exponents 
x = A[0]*np.sin(t*f[0])*np.exp(-d[0]*t) + A[1]*np.sin(t*f[1])*np.exp(-d[1]*t)
y = A[2]*np.sin(t*f[2])*np.exp(-d[2]*t) + A[3]*np.sin(t*f[3])*np.exp(-d[3]*t)

# plot them!
plt.plot(x,y,'k',linewidth=.1)
plt.axis('off')
plt.show()
```

```python
# what are the elementary components of the plots
# we have a tapering wave as attributable the the exponent of d[0]
plt.plot(np.sin(t*f[0])*np.exp(-d[0]*5*t),linewidth=.5)
plt.show()
```
